```python
from instascrape import Profile 
cr = Profile('cristiano')
cr.scrape()
recent_posts = cr.get_recent_posts()
recent_posts
```

```python
import pandas as pd 

posts_data = [post.to_dict() for post in recent_posts]
posts_data
```

```python
posts_df = pd.DataFrame(posts_data)
print(posts_df[['upload_date', 'comments', 'likes']])
```

```python
egm = posts_df['likes'] + posts_df['comments']
posts_df['engagement'] = egm
```

```python
followers = 253487386
engagement_rate = posts_df['engagement'].mean() / followers
print(format(engagement_rate,'.2%'))
```

```python
import matplotlib.pyplot as plt 

plt.style.use('seaborn-darkgrid')      # Stylistic change

plt.scatter(posts_df.upload_date, posts_df.engagement)  # Plot the data
plt.xlabel('Upload Date')              # Write labels
plt.ylabel('Engagement')
plt.title('@cristiano Engagement per Post')
plt.show()                             # Show graph 
```
