```python
import requests
url = 'https://belowcepek.com/collections/all'
response = requests.get(url)
response.text
```

```python
import pandas as pd
data = pd.DataFrame(columns = ['nama_produk', 'harga_normal', 'harga_diskon', 'gambar', 'deskripsi'])
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text, 'lxml')
items = soup.find_all("div", attrs={"class": "grid__item small--one-half medium-up--one-fifth"})#soup.find_all('div', class_='grid__item.small--one-half.medium-up--one-fifth')

for i in items:
    linki = i.find("a").get('href')
    response1 = requests.get('https://belowcepek.com'+linki)
    soup1 = BeautifulSoup(response1.text, 'lxml')
    
    #nama
    nama = i.find("div", attrs={"class":'product-card__name'}).text
    print(nama)
    #harga
    harga_normal = i.find("s", attrs={"class":'product-card__regular-price'})
    harga_diskon = soup1.find("span", attrs={"id":'ProductPrice-product-template'}).get('content')
    if (harga_normal==None):
        harga_normal = None
        harga_diskon = None
        print('soldout')
    else:
        harga_normal = harga_normal.text
    print(harga_normal)
    print(harga_diskon)
        
    #gambar
    gambar = i.find("img").get('src')
    print(gambar)
    
    #desc
    desc = soup1.find("div", attrs={"class":'rte product-single__description'}).text
    print(desc)
    
    data = data.append({'nama_produk':nama, 'harga_normal':harga_normal, 'harga_diskon':harga_diskon, 'gambar':gambar, 'deskripsi':desc}, ignore_index=True)
```

```python
def getData(response, data):
    soup = BeautifulSoup(response.text, 'lxml')
    items = soup.find_all("div", attrs={"class": "grid__item small--one-half medium-up--one-fifth"})#soup.find_all('div', class_='grid__item.small--one-half.medium-up--one-fifth')

    for i in items:
        linki = i.find("a").get('href')
        response1 = requests.get('https://belowcepek.com'+linki)
        soup1 = BeautifulSoup(response1.text, 'lxml')

        #nama
        nama = i.find("div", attrs={"class":'product-card__name'}).text
        print(nama)
        #harga
        harga_normal = i.find("s", attrs={"class":'product-card__regular-price'})
        harga_diskon = soup1.find("span", attrs={"id":'ProductPrice-product-template'}).get('content')
        if (harga_normal==None):
            harga_normal = None
            harga_diskon = None
            print('soldout')
        else:
            harga_normal = harga_normal.text
        print(harga_normal)
        print(harga_diskon)

        #gambar
        gambar = i.find("img").get('src')
        print(gambar)

        #desc
        desc = soup1.find("div", attrs={"class":'rte product-single__description'}).text
        print(desc)

        data = data.append({'nama_produk':nama, 'harga_normal':harga_normal, 'harga_diskon':harga_diskon, 'gambar':gambar, 'deskripsi':desc}, ignore_index=True)
        
    return data
```

```python
pages = soup.select('div.pagination>span.page>a')
pnum =[]
for page in pages:
    num = page.text
    pnum.append(int(num))

print(max(pnum))

for i in range(2, max(pnum)+1):
    print(url)
    response = requests.get(url+'?page='+str(i))
    data = getData(response, data)
#url = page.get('href')
#print(url)
#response = requests.get('https://belowcepek.com'+url)
#data = getData(response, data)
```
