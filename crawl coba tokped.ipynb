```python
import requests
header = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}
url = "https://www.tokopedia.com/find/sensi-convex"
response = requests.get(url, headers=header)
response.text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text, 'lxml')
pro = soup.select('div.css-1g20a2m>a')
for pr in pro:
    link = pr.get('href')
    
    response1 = requests.get(link, headers=header)
    soup1 = BeautifulSoup(response1.text, 'lxml')
    nama = soup1.find("h1", attrs={"class":'css-v7vvdw'})
    print(nama.text)
    
    harga = soup1.find("div", attrs={"class":'price'})
    print(harga.text)
    
    desc = soup1.find('div', attrs={'data-testid':"lblPDPDescriptionProduk"})
    print(desc.text)
    
    terjual = soup1.find('div', attrs={'data-testid':"lblPDPDetailProductSoldCounter"})
    print(terjual)
    
    
    print('\n')
```
